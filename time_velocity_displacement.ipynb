```python
import matplotlib.pyplot as plt
%matplotlib inline
```

# Time vs Velocity Curve at a=1 & a=2

```python
t = list(range(101))
v1 = [0]
v2 = [0]
for i in range(1,101):
    u = v1[len(v1)-1]
    v1.append(u+i)
for i in range(1,101):
    u = v2[len(v2)-1]
    v2.append(u+2*i)
plt.plot(t,v1, label='a = 1')
plt.plot(t,v2, label='a = 2')
plt.title('Time vs Velocity')
plt.xlabel('Time(sec)')
plt.ylabel('Velocity')
plt.legend()
```

# Time vs Displacement Curve at a=1 & a=2

```python
s1 = []
s2 = []
for i, j in zip(t, v1):
    s1.append(j*i+i**2/2)
for i, j in zip(t, v2):
    s2.append(j*i+i**2)
plt.plot(t, s1, label='a = 1')
plt.plot(t, s2, label='a = 2')
plt.title('Time vs Displacement')
plt.xlabel('Time(sec)')
plt.ylabel('Displacement')
plt.legend()
```
